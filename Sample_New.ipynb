```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

# Sample DataFrame
df = pd.read_csv('Airbnb_data.csv')

label_encoder = LabelEncoder()
df['room_type_encoded'] = label_encoder.fit_transform(df['room_type'])

# Check the encodingAirbnb_data

# 2. Apply StandardScaler to the encoded 'room_type' column
scaler = StandardScaler()
df['room_type_scaled'] = scaler.fit_transform(df[['room_type_encoded']])

# 5. Encoding Categorical Variables
df = pd.get_dummies(df, columns=['address.location.is_location_exact'])


# 6. Feature Engineering (example: date-time features)
df['first_review'] = pd.to_datetime(df['first_review'], errors='coerce')


# 7. Handling Outliers
# Example: Capping outliers using IQR
Q1 = df['number_of_reviews'].quantile(0.25)
Q3 = df['number_of_reviews'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['number_of_reviews'] >= lower_bound) & (df['number_of_reviews'] <= upper_bound)]


```
